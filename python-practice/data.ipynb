## 30일 숙제

```python
import pandas as pd

df = pd.read_html('https://movie.datamotion.co.kr/Boxoffice')[1]
df.columns = ['order', 'title', 'world_revenue', 'america_revenue', 'per', 'cost']
df.drop(index=0, inplace=True)
df.drop(columns=['order'], inplace=True)
df['year'] = df['title'].str[-5:-1].astype(int)
df.title = df.title.str[:-7]
df.world_revenue = df.world_revenue.str.replace(',', '').str.replace('$', '').astype(int)
df.america_revenue = df.america_revenue.str.replace(',', '').str.replace('$', '').astype(int)
df.per = round(df.per.str.replace('%', '').astype(float) / 100, 2)
df.cost = df.cost.str.replace(',', '').str.replace('$', '').astype(int)
df = df[['title', 'year', 'world_revenue', 'america_revenue', 'per', 'cost']]

df
```
