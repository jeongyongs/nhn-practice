```python
s_x = []
s_y = []
c_x = []
c_y = []

for i in range(50000):
    x = random.random() * 2
    y = random.random() * 2
    result = (x - 1) ** 2 + (y - 1) ** 2

    if result <= 1:
        c_x.append(x)
        c_y.append(y)
        continue

    s_x.append(x)
    s_y.append(y)

plt.figure(figsize=(5, 5))
plt.scatter(s_x, s_y, s=1, c="#432414")
plt.scatter(c_x, c_y, s=1, c="black")
print(4 * len(c_x) / (len(s_x) + len(c_x)))
```
